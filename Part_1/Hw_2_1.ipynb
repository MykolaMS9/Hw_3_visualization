```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96'
data_ = pd.read_html(url,decimal=',', thousands='.')
data = data_[20]
data.head(28)
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
NANcount = {}
colomn = list(data)
for col in colomn:
    NANcount[col] = (sum(data[col].isnull()))
NANcount
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
data_without_country = data.drop(27)
data_without_country
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
colomn_means = data[['1950', '1960', '1970', '1990', '1992', '1994', '1996', '2000', '2004', '2008', '2012', '2014']].mean()
colomn_means
```

```python
new_data = data.fillna(colomn_means)
new_data
```

Отримайте список регіонів, де рівень народжуваності у 2014 році був вищим за середній по Україні

```python
everage2014 = new_data.describe()['2014']['mean']
everage2014
```

```python
data_2014 = new_data[['регіон', '2014']]
data_2014[data_2014['2014'] > everage2014]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
data_2014.agg({'2014': ['max', 'idxmax']})
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2014 році

```python

plt.bar(data_2014.index, data_2014['2014'])

plt.xlabel("Регіон", fontsize="small", color="midnightblue")
plt.ylabel("Середнє значення", fontsize="small", color="midnightblue")
plt.title("Середнє значення народжувансті в регіонах за 2014 рік", fontsize=15)
```
