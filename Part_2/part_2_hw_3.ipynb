```python
def point_on_triangle(a1, a2, a3):
    """
    Random point on the triangle with vertices a1, a2 and a3.
    """
    x, y = random.random(), random.random()
    q = abs(x - y)
    s, t, u = q, 0.5 * (x + y - q), 1 - 0.5 * (q + x + y)
    return (
        s * a1[0] + t * a2[0] + u * a3[0],
        s * a1[1] + t * a2[1] + u * a3[1],
    )
```

```python
def add_point(x,y,point):
    x = np.append(x, point[0])
    y = np.append(y, point[1])
    return x, y

def serpinskiy(a1, a2, a3, x, y):
    point = np.array(point_on_triangle(a1, a2, a3))
#     x, y = add_point(x, y, point)
    
    for _ in range(10000):
        apex = random.randint(0, 2)
        point = np.array([np.mean([x[apex], point[0]]), np.mean([y[apex], point[1]])])
        x, y = add_point(x, y, point)
    return x, y
```

```python
triangle_length = 10
a1 = np.array([0,0])
a2 = np.array([triangle_length,0])
a3 = np.array([triangle_length/2,np.abs(triangle_length*np.sin(60))])
```

```python
x = np.array([a1[0], a2[0], a3[0]])
y = np.array([a1[1], a2[1], a3[1]])

x, y = serpinskiy(a1, a2, a3, x, y)

plt.scatter(x,y, s=1)
```
